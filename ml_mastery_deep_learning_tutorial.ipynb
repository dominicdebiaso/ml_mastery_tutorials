#### Lession 01: Introduction to Theano 

```python
import theano
from theano import tensor
```

```python
a = tensor.dscalar()
b = tensor.dscalar()
z = tensor.dscalar()
```

```python
c = (a + b) * z
```

```python
f = theano.function([a,b,z], c)
```

```python
result = f(1.5, 2.5, 10)
```

```python
print(result)
```

#### Lesson 02: Introduction to TensorFlow

```python
import tensorflow as tf
```

```python
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
```

```python
add = tf.add(a,b)
```

```python
c = sess.run(add, feed_dict=binding)
```

#### Lesson 05: Your First Neural Network in Keras

```python
import pandas as pd
import numpy
from keras.models import Sequential
from keras.layers import Dense

seed = 7
numpy.random.seed(seed)
```

```python
df = pd.read_csv('/Users/dominicdebiaso/Development/datasets/pima-indians-diabetes.data.csv', header=None)
```

```python
X = df.loc[:,0:7].values
y = df.loc[:,8].values
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, nb_epoch=150, batch_size=10)
```

```python
scores = model.evaluate(X, y)
```

```python
print(zip(model.metrics_names, [(i*100).round() for i in scores]))
```

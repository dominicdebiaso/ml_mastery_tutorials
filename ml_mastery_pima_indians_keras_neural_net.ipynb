```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv('/Users/dominicdebiaso/Development/datasets/pima-indians-diabetes_data.csv', header=None)
```

```python
X = df.loc[:,0:7].values
y = df.loc[:,8].values
```

```python
## Define model
# Create sequence of layers one at a time until network topology is satisfactory
model = Sequential()
# first layer has 12 neurons, 8 input variables, uniform distribution
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
# hidden layer
model.add(Dense(8, init='uniform', activation='relu'))
# output layer has 1 neuron to predict class as sigmoid to ensure output is between 0 and 1
model.add(Dense(1, init='uniform', activation='sigmoid')) 

## Compile model
# logarithmic loss for binary classification using gradient descent and 
# report classification accuracy
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

## Fit model
# run thought a number of iterations (ie. epochs) and 
# number of evaluated instances prior to a weight update
model.fit(X, y, nb_epoch=150, batch_size=10)

## Evaluate model
scores = model.evaluate(X, y)
print(model.metrics_names[1], scores[1]*100)
```

```python
## Make predictions
preds = model.predict(X)
pd.DataFrame(preds.round(), columns={'preds'}).preds.value_counts()
```
